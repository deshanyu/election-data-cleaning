```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
dfs = []
usecols = ['CONTRIBUTOR', 'STREET1', 'CITY', 'STATE', 'ZIP','EMP NAME', 'OCCUPATION NAME', 'RECIPIENT NAME','RECIPIENT LOCATION',
'RECIPIENT ELECTION TYPE', 'RECIPIENT ELECTION YEAR','CONT AMT','CONT DATE', 'CONTRIBUTOR TYPE']
for f in all_filenames:
    df = pd.read_csv(f,usecols = usecols)
    df['filename'] = f
    dfs.append(df)
# combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
combined_csv = pd.concat(dfs)
combined_csv.head()
```

```python
combined_csv.shape
```

```python
new_names = {'STATE': 'contributorstate',
'STREET1': 'contributoraddress',
'CITY': 'contributorcity',
'RECIPIENT LOCATION': 'city',
'RECIPIENT ELECTION YEAR': 'year',
'RECIPIENT ELECTION TYPE': 'type',            
'RECIPIENT NAME': 'candidate',
'CONT DATE':  'contributiondate',
'CONTRIBUTOR': 'contributorname',
'ZIP':  'contributorzip',
'CONTRIBUTOR TYPE': 'contributiontype',
'OCCUPATION NAME': 'contributoroccupation',
'EMP NAME': 'contributoremployer',
'CONT AMT': 'contributionamt'}
combined_csv.rename(columns = new_names, inplace = True)
```

```python
combined_csv['state'] = 'NJ'
```

```python
combined_csv['contributorzip'] = combined_csv['contributorzip'].astype(str)
# df['zipcode'] = df.zipcode.astype(str)
```

```python
combined_csv.info()
```

```python
def find_lastname(full_name):
    
    last_name=full_name.split(" ")[0]
    
    return last_name
     
combined_csv['candidatelastname'] = combined_csv['candidate'].apply(find_lastname)
combined_csv.head()
```

```python
def zipconvert(x):
    if len(x) == 4:
        return "0" + x
    return x
combined_csv['contributorzip'] = combined_csv['contributorzip'].apply(zipconvert)
combined_csv.head()
```

```python
#export to csv
combined_csv.to_csv( "nj_csv.csv", index=False, encoding='utf-8-sig')
```
